### Eliminar valores duplicados

```python
# Eliminar duplicados

hr_raw_data.drop_duplicates(inplace=True) 
```

### Homogeneización de los datos

```python
nuevas_columnas = {col: col.lower().replace('.','_') for col in hr_raw_data.columns}
nuevas_columnas
```

```python
hr_raw_data.rename(columns = nuevas_columnas, inplace=True)
hr_raw_data.columns
```

```python
columnas_a_cambiar = ["businesstravel", "department", "educationfield", "jobrole", "maritalstatus", "standardhours", "attrition", "roledepartament"]
```

```python
for col in columnas_a_cambiar:
    hr_raw_data[col] = (
    hr_raw_data[col]
    .str.strip()            # quita espacios al inicio y fin
    .str.lower()            # pasa a minúsculas
    .str.replace(" - ", "_", regex=False)
    .str.replace("-", "_", regex=False)
    .str.replace("&", " and ", regex=False) 
    .str.replace("/", "_", regex=False)   
    .str.replace(" ", "_", regex=False)
    .str.replace(r"_+", "_", regex=True)   
    .str.strip("_")                          
)
```

```python
hr_raw_data
```

#### Convertir a numérico

```python
convertir_a_numerico = ["monthlyincome", "age", "monthlyrate", "totalworkingyears", "salary", "performancerating", "worklifebalance"]


fconvertir_a_numerico = [
    "monthlyincome", 
    "age", 
    "monthlyrate", 
    "totalworkingyears", 
    "salary", 
    "performancerating", 
    "worklifebalance"
]

for col in convertir_a_numerico:
    if col == "monthlyincome":
        # Directo a numérico, sin convertir a string
        hr_raw_data[col] = pd.to_numeric(hr_raw_data[col], errors="coerce")
    else:
        # Solo limpiar si la columna es object (texto con símbolos)
        if hr_raw_data[col].dtype == "object":
            hr_raw_data[col] = (
                hr_raw_data[col]
                .str.replace("$", "", regex=False)
                .str.replace(",", ".", regex=False)
                .str.strip()
            )
        # Convertir a numérico
        hr_raw_data[col] = pd.to_numeric(hr_raw_data[col], errors="coerce")
```
